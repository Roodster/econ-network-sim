```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random 

import sys
sys.path.append('./components')
```

```python
from components.clusters import Cluster
```

```python
group_ids = [
    'YELLOW',
    'RED',
    'BLUE',
    'GREEN',
    'ORANGE',
    'PINK',
    'PURPLE',
]
```

```python
c = Cluster(initial_value=10, group_id=group_ids.pop(), brownian_s0=10)
c.create_cluster(size=5)

print(c.cluster)

clusters = np.array([Cluster(initial_value=random.randint(5, 10), group_id=group_ids.pop(), brownian_s0=10) for _ in range(5)])
for cluster in clusters:
    cluster.create_cluster(size=5)

for idx, cluster in np.ndenumerate(clusters):
    
    print(cluster)
```

```python
plt.figure(figsize=(9,4))
for idx, cluster in np.ndenumerate(clusters):
    plt.plot(cluster.brownian)
plt.legend(['Scenario-'+str(i) for i in range(1,6)],
            loc='upper left')
plt.hlines(y=100,xmin=0,xmax=100,
            linestyle='--',color='k')
plt.show()
```
